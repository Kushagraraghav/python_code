```python
import numpy as np
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
list3 = [11, 12, 13, 14, 15]
arr = np.array([list1, list2, list3])
print(arr[1:2,1:3])
print("Original array:")
print(arr)
new_row = [20, 21, 22, 23, 24]
arr = np.insert(arr, 1, new_row, axis=0)
print("\nAfter inserting values:")
print(arr)
print(arr.reshape(2,10))
print(arr[1:2,1:3])
print(arr[ :1 ,3 :])
```
